## Demonstration of Open Chemistry integration of Avogadro in JupyterLab

```python
import avogadro as avo
import openchemistry as oc
import json
```

Once you have imported the basic Python modules we load a file

```python
m = avo.core.Molecule()
f = avo.io.FileFormatManager()
# This API will change a little soon, readFile -> read_file, writeString -> write_string...
# Read a file using Avogadro's wrapped API
f.readFile(m, 'h2o-restricted.fchk')
# Then write it to CJSON, we plan on removing the need to do this soon.
cjson_str = f.writeString(m, 'cjson')
cjson = json.loads(cjson_str)
```

We will be improving the interface, but for now remove the vibrations key as we are not interested in that for this example.

```python
print("electronCount = ", cjson['orbitals']['electronCount'])
if 'vibrations' in cjson:
    del cjson['vibrations']
mol = oc.load(cjson)
```

Show the orbital, note that this is calculated using the Avogadro 2 libraries from the output in the CJSON file. You can use numeric orbital numbers, or homo/lumo.

```python
mol.structure.show()
```

```python
mol.orbitals.show(mo='lumo', iso=0.09)
```

```python
mol.orbitals.show(mo='homo', iso=0.1)
```
